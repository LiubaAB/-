```python
# Для наглядности установлю границы по макс и мин датам из adgrants_charityhospital и построю гистограмму по столбцу Первый платеж
(
    sns.displot(
        data=clients[(clients['Последний платеж'] >= '2023-01-15') & (clients['Последний платеж'] <= '2023-07-19')],
        x='Последний платеж', kde=True, height=5, aspect=5)
)
plt.title('Последний платёж, таблица clients')
plt.xlabel('Дата')
plt.ylabel('Количество платежей')
plt.show()
(
    sns.displot(
        data=donors[(donors['Дата последнего пожертвования'] >= '2023-01-15') & (donors['Дата последнего пожертвования'] <= '2023-07-19')],
        x='Дата последнего пожертвования', kde=True, height=5, aspect=5)
)
plt.title('Последний платёж, таблица donors')
plt.xlabel('Дата')
plt.ylabel('Количество платежей')
plt.show()
# Посчитаю число показов в разбивке по датам
times_shown = (
    adgrants_charityhospital.groupby('Дата', as_index=False).agg({'Показы':'sum'})
    .sort_values(by='Показы', ascending=False)
    .reset_index(drop=True)
)
# Построю линейный график по суммарным показам в разбивке по датам для сравнения пиков по платежам и показам
times_shown.plot(x='Дата', y='Показы', figsize=(20,5), grid=True)
plt.title('Показы (таб adgrants_charityhospital)')
plt.xlabel('Дата')
plt.ylabel('Сумма показов')
plt.show()
```

Пики по последним платежам не совпадают с пиками по показам. Посмотрю, какие кампании проходили в пиковый период по показам (середина-конец марта 2023 года), который находится близко к пикам по первым платежам (март 2023 года в donors и апрель 2023 года в clients).

```python
(
    adgrants_charityhospital[(adgrants_charityhospital['Дата'] >= '2023-03-15') & (adgrants_charityhospital['Дата'] <= '2023-04-01')]
     ['Кампания'].unique()
)
```

По косвенным признакам можно предположить, что эти кампании наиболее успешны. Ранжирую их по количеству показов (единственная метрика, по которой у нас есть данные), добавлю данные с началом и окончанием кампании

```python
# Добавлю столбец с началом кампании
start_date = (
    adgrants_charityhospital[(adgrants_charityhospital['Дата'] >= '2023-03-15') & (adgrants_charityhospital['Дата'] <= '2023-04-01')]
    .groupby('Кампания', as_index=False)
    .agg({'Показы':'sum', 'Дата':'min'})
    .rename(columns={'Дата': 'Дата начала кампании'})
    .sort_values(by='Показы', ascending=False)
    .reset_index(drop=True)
)
# Добавлю столбец с окончанием кампании
end_date = (
    adgrants_charityhospital[(adgrants_charityhospital['Дата'] >= '2023-03-15') & (adgrants_charityhospital['Дата'] <= '2023-04-01')]
    .groupby('Кампания', as_index=False)
    .agg({'Показы':'sum', 'Дата':'max'})
    .rename(columns={'Дата': 'Дата завершения кампании'})
    .sort_values(by='Показы', ascending=False)
    .reset_index(drop=True)
)
# Объединю таблицы
campaigns = start_date.merge(end_date, how='left', on=['Кампания', 'Показы'])
```

На этом графике мы видим кампании, даты начала которых близки к датам пика по количеству пожертвований. На основании близости пиков можно предположить, что пожертвования были сделаны благодаря этим кампаниям. Эти кампании можно рассматривать как ниболее эффективные.
